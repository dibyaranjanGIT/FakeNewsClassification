```python
stop = set(stopwords.words('english'))
```

```python
def remove_string_special_characters(text): 
    
    #for i in range(0, 23480):
            
    # removes special characters with ' ' 
    stripped = re.sub('[^a-zA-z\s]', '',str(text))
    stripped = re.sub('_', '', stripped) 
    
    # Change any white space to one space 
    stripped = re.sub('\s+', ' ', stripped) 
    
    # Remove start and end white spaces 
    stripped = stripped.strip() 
    
    return stripped.lower()
            
```

```python
def remove_stopwords(text):
    final_text = []
    for i in text.split():
        if i.strip().lower() not in stop:
            final_text.append(i.strip())
    return " ".join(final_text)
```

```python
def clean_text (text):
    text_r = remove_string_special_characters(text)
    final_text = remove_stopwords(text_r)
    return final_text
```

```python
df ['cleaned_text'] = df['text_title'].apply ( lambda x : clean_text(x)) 
```

```python
df.head()
```

```python
# Applying CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=5000,ngram_range = (1,3)) 
X_v = vectorizer.fit_transform(df['cleaned_text'])
#features = (vectorizer.get_feature_names()) 
```

```python
print("\n\nFeatures : \n", vectorizer.get_feature_names())
```

```python
#vectorizer.vocabulary_
```

```python
y = df['category']
```

```python
import pickle
```

```python
# creating a pickle file for converting the data into an Array
pickle.dump(vectorizer,open('CV_tranformer.pkl','wb'))
```

```python
X_v.shape
```

```python
y.shape
```

```python
## Divide the dataset into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_v, y, test_size=0.3, random_state=0)
```

### Support Vector Classifier Algorithm

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10,100],  
              #'gamma': [1, 0.1, 0.01], 
              'kernel': ['rbf','linear']}  
```

```python
svc = SVC()
```

```python
grid = GridSearchCV(svc, param_grid, refit = True, verbose = 3 ,n_jobs = 4) 
```

```python
grid.fit(X_train, y_train) 
```

```python
#To know to best parameters on which the model is trained on
grid.best_params_
```

```python
import pickle
```

```python
# creating a pickle file for converting the data into an Array
pickle.dump(grid,open('GridClassifier.pkl','wb'))
```

### Evaluation 

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
pred = grid.predict(X_test)
score = accuracy_score(y_test, pred)
```

```python
pred
```

```python
print("Accuracy:   %0.3f" % score)
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(pred,y_test))
```
